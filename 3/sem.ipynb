## Работа со списками

Дана матрица 3х2, пример:
```
matrix = [
    [1, 2],
    [5, 6],
    [9, 10],
]
```

Необходимо написать функцию, которая будет её транспонировать, т.е. приводить к виду
```
matrix = [
    [1, 5, 9],
    [2, 6, 10],
]
```

## Работа со множествами

Даны 2 множества, к примеру

s1 = {1, 2, 3, 4, 5, 6, 7}
и s2 = {5, 6, 7, 8, 9}

Требуется:

- найти уникальные элементы из обоих множеств
- найти общие элементы для обоих множеств

## Работа со словарями

Решите задачу со множествами используя словари (элементы - ключи). 

Входные данные при этом поступают в виде списков (все элементы уникальны), к примеру:

s1 = [1, 2, 3, 4, 5, 6, 7]
и s2 = [5, 6, 7, 8, 9]
