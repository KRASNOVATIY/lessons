## Функциональный стиль

Немного теории

2 наиболее распространённые парадигмы программирования:

    1. Императивнае. Отвечает на вопрос "Как?" - какие действия хотите выполнить
    2. Декларативнае. Отвечает на вопрос "Что?" - что вы хотите получить

Языки программирования поддерживают декомпозицию проблем несколькими различными способами:

    1. Большинство языков программирования являются процедурными (императивный подход): программа представляет собой набор инструкций, которые сообщают компьютеру, что делать с входными данными.

    2. В декларативных языках вы пишете спецификацию, которая описывает решаемую проблему. SQL - это декларативный язык

    3. Объектно-ориентированные программы (императивный подход) управляют наборами объектов. Объекты имеют внутреннее состояние и поддерживают методы, которые каким-либо образом изменяют это внутреннее состояние.

    4. Функциональное программирование (декларативный подход) разбивает проблему на набор функций. В идеале функции принимают только входные данные, производят выходные данные и не имеют никакого внутреннего состояния (которое влияет на выходные данные).
    

Говоря о ФП, сразу следует подчеркнуть, что программирование через функции далеко не всегда ФП, чаще всего это всего лишь процедурный стиль программирования. 

Python является мультипарадигмальным, однако использование чистого декларативного стиля является плохой практикой... перегибом, крайностью. Вместо этого, на практике, применяется функциональный стиль: объекты используются для хранения состояний, при этом программа предоставлет функциональный интерфейс


````python
x = 2
    
# Пример нормального управления потоком выполнения
result = 0
if x <= 1: 
    result = 10 
elif 10 > x > 1: 
    result = 20  
else: 
    result = 30
    
print(result)

# Пример короткого (функционального) управления
(x <= 1 and 100) or (10 > x > 1 and 200) or (300)
````

Базовые элементы ФП в Python — функции `map()`, `filter()`, `iter()`, `zip()`, оператор `lambda`, генераторы последовательностей и выражения-генераторы. Продвинутые элементы - библиотеки `operator`, `functools`, `itertools`

`map()` позволяет заменить цикл обработки

````python
items = [-1, -2, -3]

# ПП
results = []
for i in items:
    results.append(abs(i))
print(results)

# ФП
print(list(map(abs, items)))

# выражения генераторы
print([abs(x) for x in items])
````

`filter()` позволяет заменить цикл обработки с условием

````python
items = [1, -2, 3]

# ПП
results = []
for i in items:
    if i > 0:
        results.append(i)
print(results)

# ФП
print(list(filter(lambda x: x > 0, items)))  # чтобы это было чистое ФП, пришлось использовать lambda

# выражения генераторы
print([x for x in items if x > 0])
````

## Итераторы и генераторы

### итераторы

Итератор - это объект, представляющий поток данных; этот объект возвращает данные по одному элементу за раз.

Итератор Python должен поддерживать вызываемый метод `__next__()`, который не принимает аргументов и всегда возвращает следующий элемент потока. Если в потоке больше нет элементов, `__next__()` должен вызвать исключение `StopIteration`.

Итераторы могут быть бесконечными.

`iter(object)` - пытается вернуть итератор, который содержит элементы объекта object

````python
t = 1, 2, 3
iterator = iter(t)
while True:
    print(next(iterator))
````

````python
t = 1, 2, 3
iterator = iter(t)
for i in iterator:  # заметим, тут исключения StopIteration не будет
    print(i)
````

````python
for i in iterator:  # повторный запуск ничего не вернёт! Итератор пуст
    print(i)
````

````python
# итератор может быть материализован как список или кортеж
t = 1, 2, 3
iterator = iter(t)
# print(list(iterator))
print(tuple(iterator))
````

````python
# можно распаковать итератоор
t = 1, 2, 3
iterator = iter(t)
a, b, c = iterator
a, b, c
````

### генераторы

Генераторы - это особый класс функций, которые упрощают задачу написания итераторов. 

Ключевое слово - `yield`

Обычные функции вычисляют значение и возвращают его. Когда интерпретатор доходит до ключевого слова `return`, выполнение функции полностью прекращается.

Генераторы возвращают итератор, который возвращает поток значений. Когда интерпретатор доходит до ключевого слова `yield`, программа приостанавливает выполнение функции и возвращает значение в итерируемый объект.

В отличии от списка, генератор не хранит в памяти все элементы, он вычисляет их один за другим, в процессе итерации, удерживает только одно значение — то, которое он возвращает 

````python
def gen(start, stop):
    for i in range(start, stop):
        yield i
        
g = gen(7, 10)
g
````

````python
next(g)
````

````python
list(g)
````

````python
next(g)
````

Генераторы способны не только возвращать данные, но и принимать их 
```python
var = (yield i)
```

````python
def counter(maximum):
    i = 0
    while i < maximum:
        val = (yield i)
        if val is not None:
            i = val
        else:
            i += 1
````

````python
c = counter(20)
next(c)
````

````python
next(c)
````

````python
c.send(18)  # передаём значение в генератор
````

````python
next(c)
````

````python
next(c)
````

Что это даёт:
1. генераторы могут быть как производителями, так и потребителями данных
2. генераторы также становятся сопрограммами (англ. coroutine) - более обобщенной формой подпрограмм. Подпрограммы вводятся в одной точке и завершаются в другой точке (начало функции и return оператор), но сопрограммы можно вводить, завершать и возобновлять во многих разных точках (yield операторах).

## Рекурсия

Рекурсивная функция – это функция, которая вызывает сама себя. 

````python
def recursive():
    recursive()

recursive()
````

Чтобы стек вызовов не переполнялся, или не была достигнута максимальная глубина рекурсии, в каждой рекурсивной функции всегда должны быть предусмотрены два случая:

    - граничный, при котором функция завершает работу и возвращает данные в основную программу.
    - рекурсивный, при котором функция продолжает вызывать себя.

````python
def eat_list(lst):
    if not lst:
        return
    print(lst.pop())
    return eat_list(lst)
    
eat_list([10, 8, 6, 5])
````

Важно помнить что рекурсивные функции работают медленнее обычных и занимают больше памяти, поэтому их стоит применять только тогда, когда решить задачу без рекурсии сложно (невозможно). Кроме того рекурсия ухудшает читаемость кода

````python
# сравним скорость выполнения рекурсивной и итеративной функции, оценим читаемость и понятность кода 

from timeit import timeit

def fib_it(n):
    if n == 1:
        return [1]
    if n == 2:
        return [1, 1]
    fibs = [1, 1]
    for _ in range(2, n):
        fibs.append(fibs[-1] + fibs[-2])
    return fibs

setup_code_iter = 'from __main__ import fib_it'
stmt_iter = 'fib_iter(15)'
print('Время выполнения итеративной функции: ', timeit(setup=setup_code_iter, stmt=stmt_iter, number=20000))

def fib_recursive(n):
    if(n <= 1):
        return n
    else:
        return(fib_recursive(n-1) + fib_recursive(n-2))
    
setup_code_rec = 'from __main__ import fib_rec'
stmt_rec = 'fib_recursive(15)'
print('Время выполнения рекурсивной функции: ', timeit(setup=setup_code_rec, stmt=stmt_rec, number=20000))
````

Ради справедливости стоит отметить что конкретно последовательность фибоначи плохо вычисляет через рекурсию,

## Система исключений и assert

### assert

Утверждения (assertions) — это инструкции, которые «утверждают» определенный случай в программе.

Синтаксис:

`assert <condition>[, message]`

````python
assert True, 'message'
````

````python
assert False
````

Применяются утверждения для простейшего тестирования

````python
def my_plus(a, b):
    return a + b

def test_mu_plus(fun_args, result):
    assert my_plus(*fun_args) == result, f'sum of {fun_args} != {result}'
````

````python
test_mu_plus((1, 2), 3)
````

````python
test_mu_plus((1, 2), 4)
````

### exceptions

Исключениями (exceptions) в языках программирования называют проблемы, возникающие в ходе выполнения программы, которые допускают возможность дальнейшей ее работы в рамках основного алгоритма. Примеры исключений:
- деление на ноль
- невозможность читать / писать в файл
- проблемы с сетью
- завершение итератора
- отсутсвие ключа в словаре
- недопустимые аргументы функции

Существует довольно большое количество встроенных типов исключений в языке Python, все они составляют определенную иерархию, в корне которой находится BaseException

### обработка исключений

Обработка исключений нужна для того, чтобы приложение не завершалось аварийно каждый раз, когда возникает исключение.

Синтаксис:
```python
try:
    result = statement_try  # действия которые могут возбуждать исключения
except [ExceprionType1, ... ExceprionTypeK as exception_alias1]:  # перехват исключений 
    statement_e1
except [ExceprionTypeM, ... ExceprionTypeN as exception_alias2]:  # перехват исключений, не перехваченных выше 
    statement_e2
    [raise ExceprionTypeU from exception_alias2]  # возбуждаем исключение ExceprionTypeU из возникшего (для сохранения стека)
    [pass] # можно ничего не делать с исключением (плохая практика)
...
else:  # действия, которые нужно совершить если исключения не возникли
    statement_try_success(result)
finally:  # действия, которые нужно совершить в любом случае
    statement_try_success_or_fail
```

Заметим, что если в блоке statement_e1 возникнет исключение, оно не будет перехвачено ниже

````python
try: 
    1 / 0
except ZeroDivisionError:
    1 / ''
except TypeError:
    pass
````

````python
def divide(a, b):
    try:
        return a / b
    except:
        print('Noo!')
    
divide(1, 2)
````

````python
divide(1,'2')
````

````python
def divide(a, b):
    try:
        result = a / b
    except Exception as e:
        print(e)
    else:
        return result
    finally:
        print('End')
    
divide(1, 2)
````

````python
divide(1,'2')
````

````python
def divide(a, b):
    try:
        return a / b
    except ZeroDivisionError as e:
        return 0
    except TypeError as e:
        raise ValueError('Some context') from e
        
divide(1, 0)
````

````python
divide(1, '1')
````

### возбуждение исключений

Для принудительной генерации исключения используется инструкция raise. 

Синтаксис:

`raise ExceptionName[(*exc_args)]`

````python
def divide(a, b):
    if not isinstance(a, int) or not isinstance(b, int):
        raise ValueError
    return a / b

divide(1, 1.1)
````

### пользовательские исключения

В Python можно создавать собственные исключения. Такая практика позволяет увеличить гибкость процесса обработки ошибок в рамках той предметной области, для которой написана программа.

Для реализации собственного типа исключения необходимо создать класс, являющийся наследником от одного из классов исключений.

````python
class NegativeValueException(Exception):
   ...

def divide(a, b):
    if a < 0 or b < 0:
        raise NegativeValueException('Ну зачем же нам тут отрицательное!')
    return a / b

divide(1, 2)
````

````python
divide(1, -2)
````
